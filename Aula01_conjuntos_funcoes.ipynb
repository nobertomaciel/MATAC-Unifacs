```python
# Operações com conjuntos
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print('União:', A | B)
print('Interseção:', A & B)
print('Diferença A-B:', A - B)
```

## Exercício: Modifique os conjuntos com linguagens de programação

```python
# Funções em Python
def f(x):
    return 2*x + 3

print(f(2))
print(f(-1))
```

```python
# Composição de funções
def f(x): return x + 1
def g(x): return 2*x

print('f(g(x)):', f(g(3)))
print('g(f(x)):', g(f(3)))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), label='f(x)', color='red')
plt.plot(x, g(x), label='g(x)', color='blue')
plt.plot(x, f(g(x)), label='f(g(x))', color='green')
plt.legend()
plt.show()
```

```python
# Composição de funções
def f(x): return 2*x + 1
def g(x): return x**2 + 2*x

print('f(g(x)):', f(g(3)))
print('g(f(x)):', g(f(3)))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), label='f(x)', color='red')
plt.plot(x, g(x), label='g(x)', color='blue')
plt.plot(x, f(g(x)), label='f(g(x))', color='green')
plt.legend()
plt.show()
```
